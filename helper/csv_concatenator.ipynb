```python
import pandas as pd
```

```python
file1 = "wasabi_songs_without_genre_1500200.csv"
file2 = "wasabi_songs_without_genre_second_half.csv"

dtypes = {'_id': 'str', 'position': 'Int64', 'lengthAlbum': 'str', 'lyrics': 'str', 
          'urlWikipedia': 'str', 'id_album': 'str', 'isClassic': 'bool', 'title': 'str',
          'publicationDateAlbum': 'str', 'albumTitle': 'str', 'isrc': 'str', 
          'length': 'Int64', 'explicitLyrics': 'str', 'rank': 'Int64', 
          'bpm': 'float64', 'gain': 'float64', 'preview': 'str', 'publicationDate': 'str', 
          'language': 'str', 'language_detect': 'str','name': 'str', 
          'explicit_content_lyrics': 'Int64'}
```

```python
df1 = pd.read_csv(file1, sep='\t', decimal='.', dtype=dtypes)
df2 = pd.read_csv(file2, sep='\t', decimal='.', dtype=dtypes)
```

```python
df1
```

```python
df2
```

```python
df_final = pd.concat([df1, df2], axis=0, ignore_index=True)
df_final
```

```python
df_final = df_final.drop_duplicates(ignore_index=True)
```

```python
df_final
```

```python
df_final.to_csv('wasabi_songs_without_genre_final.csv', sep='\t', encoding='utf-8', index=False)
```
